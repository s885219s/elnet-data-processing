```python
import numpy as np
import argparse
import imutils
from imutils import face_utils
import dlib
import cv2
import glob
import shutil
import os 
shape_predictor = "shape_predictor_68_face_landmarks.dat"
root = "D:/share/new_dataset/new_human_power_reorder/dirl_face/"
box_root = "D:/share/new_dataset/new_human_power_reorder/dirl_face/face_feature_points/openface/DIRL"
input_root = "D:/share/new_dataset/new_human_power_reorder/dirl_v1/origin_image/"
input_txt_root = "D:/share/new_dataset/new_human_power_reorder/dirl_v1/feature_points/"
output_root = "D:/share/new_dataset/new_human_power_reorder/dirl_face/face_image/DIRL_openface/"
output_txt_root = "D:/share/new_dataset/new_human_power_reorder/dirl_face/face_feature_points/DIRL_openface/"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(shape_predictor)
```

```python
def read_fp(txt):
    fp = np.zeros((14,2),dtype = np.int32)
    with open(txt,'r') as f:                        
        lines = f.read()                
        lines = lines.split('\n')
        lines = lines[:-1]
        for i,line in enumerate(lines):
            point = line.split(',')
            fp[i,0] = int(point[2])#/width
            fp[i,1] = int(point[3])#/height
    return fp
```

```python
def read_box(box):
    fp = np.zeros((68,2),dtype = np.int32)
    with open(box,'r') as f:                        
        lines = f.read()                
        lines = lines.split('\n')
        lines = lines[:-1]
        for i,line in enumerate(lines):
            point = line.split(',')
            fp[i,0] = int(point[0])#/width
            fp[i,1] = int(point[1])#/height
        x1 = min(fp[:,0])
        y1 = min(fp[:,1])
        x2 = max(fp[:,0])
        y2 = max(fp[:,1])
        xcen = (max(fp[:,0])+min(fp[:,0]))*0.5
        ycen = (max(fp[:,1])+min(fp[:,1]))*0.5
        width = max(max(fp[:,0])-min(fp[:,0]),max(fp[:,1])-min(fp[:,1]))*0.5
    return int(xcen-width*1.1+0.5), int(ycen-width*1.1+0.5), int(xcen+width*1.1+0.5), int(ycen+width*1.1+0.5)
```

```python
with open(root+'dirl_fileList.txt','r') as t:
    lines = t.read()
    lines = lines.split('\n')
    lines = lines[:-1]
    for line in lines:
        fn = os.path.basename(line)
        file = os.path.join(input_root,line)
        txt = os.path.join(input_txt_root,line)[:-4]+'.txt'
        box = os.path.join(box_root,line)[:-4]+'_face.txt'
        path = os.path.dirname(line)
        classify_img = os.path.join(output_root,path)
        classify_txt = os.path.join(output_txt_root,path)
        if not os.path.exists(classify_img):
            os.makedirs(classify_img)
        if not os.path.exists(classify_txt):
            os.makedirs(classify_txt)
        if fn.endswith('jpg'):
            fn = fn[:-4]+'.png'
        image = cv2.imread(file)
        x1,y1,x2,y2 = read_box(box)
        image = image[y1:y2,x1:x2]
        fp = read_fp(txt)
        fp[:,0] = fp[:,0]-x1
        fp[:,1] = fp[:,1]-y1
        cv2.imwrite(os.path.join(classify_img,fn), image)
        with open(os.path.join(classify_txt,fn)[:-4]+'.txt','w') as f:
            for i,p in enumerate(fp):
                f.write(str(i//7)+','+str(i%7)+','+ str(p[0])+','+str(p[1])+'\n')
        print(fn)
```
